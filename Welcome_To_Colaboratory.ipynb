```
import matplotlib.pyplot as plt
import matplotlib
from mpl_toolkits.mplot3d import Axes3D
```

```
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)
```

```
figure = plt.figure(1, figsize = (12, 4))
subplot3d = plt.subplot(111, projection='3d')
surface = subplot3d.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=matplotlib.cm.coolwarm, linewidth=0.1)
plt.show()
import pandas as pd
df = pd.read_csv('/content/sample_data/Au_nanoparticle_dataset.csv')
print(df.shape)
df.columns
for i in range (df.shape[0]):
  n_sum = (df['N_bulk'][i] + df['N_surface'][i])
  assert n_sum == df['N_total'][i]
  if(df.shape[0]-1) == i:
    print('all N_bulk and N_surface add up to N_total')
new_df = df[['N_total', 'N_bulk', 'N_surface', 'R_avg']]
new_df.describe()
new_df.head(20)
```

```
mean_values = new_df.mean()
print("Mean Values : ")
print(mean_values)
```

```
std_values = new_df.std()
print("\n Standard Deviation Values : ")
print(std_values)
```

```
quartiles = new_df.quantile([0.25, 0.5, 0.75])
print("\n Quartiles : ")
print(quartiles)
```

```
new_df.loc[:, ['N_total', 'N_bulk', 'N_surface', 'R_avg']].hist(layout=(1,4), bins=20, figsize=(40,10))
import seaborn as sns
import matplotlib.pyplot as plt
sns.pairplot(new_df)
plt.show()
g = sns.PairGrid(new_df)
g.map_upper(sns.histplot) #bivariate histogram
g.map_diag(sns.histplot, kde=True)
g.map_lower(sns.kdeplot)
```
